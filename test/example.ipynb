```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Load dataset
data = pd.read_csv(
  filepath_or_buffer="../penyakit-ayam.csv"
)

# Define readable feature names and class names
feature_names = [
  "Gangguan Syaraf", "Diare", "Nafsu Makan Menurun", "Merejan",
  "Tubuh Gemetar", "Bulu Kusam atau Kotor", "Bulu Sekitar Anus kotor", "Lesu"
]
class_names = [
    "Gumboro", "Tetelo", "Snot", "Infection Bronchitis",
    "Avian influenza", "Cacar Ayam", "Limfoid Leukosit", "Marek"
]

# Prepare features (X) and target (y)
X = data[['G1', 'G2', 'G3', 'G4', 'G5', 'G6', 'G7', 'G8']]
y = data['Penyakit']

# Create and train the decision tree classifier
clf = DecisionTreeClassifier(max_depth=3, random_state=42)
clf.fit(X, y)

# Visualize the decision tree
plt.figure(figsize=(15, 10))
plot_tree(clf, feature_names=feature_names, class_names=class_names, filled=True, rounded=True, fontsize=10)
plt.title("Decision Tree untuk Penyakit Ayam")
plt.show()
```
